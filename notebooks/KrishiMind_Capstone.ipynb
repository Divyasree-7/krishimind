## Project Overview
KrishiMind is a multi-agent AI system that acts as an intelligent farming companion for Indian smallholder farmers.
It combines real weather data, crop knowledge, market intelligence, and pest risk assessment into a unified conversational assistant.

### Architecture
```
User Query → Orchestrator Agent → [Weather | Crop | Market | Pest] Agents → Synthesized Response
```
### Agents
1. **Weather & Climate Agent** — Open-Meteo API, rainfall risk scoring
2. **Crop Advisory Agent** — ICAR knowledge base + Gemini reasoning
3. **Market Price Agent** — Mandi prices, MSP comparison, sell-window advice
4. **Pest & Disease Agent** — Seasonal heuristics + LLM prevention guidance
5. **Orchestrator** — Routes queries, coordinates agents, synthesizes responses
